```python
!pip install pandas-profiling --quiet

# URL of the medical charges dataset
medical_charges_url = 'https://raw.githubusercontent.com/JovianML/opendatasets/master/data/medical-charges.csv'

from urllib.request import urlretrieve
import pandas as pd

# Download the file
urlretrieve(medical_charges_url, 'medical.csv')

# Load the CSV file into a DataFrame
medical_df = pd.read_csv('medical.csv')
```

```python
# Install the required libraries (only needed if not already installed)
!pip install plotly matplotlib seaborn --quiet

# Import the necessary libraries
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Customize plot appearance
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.facecolor'] = '#00000000'
```

```python
medical_df.age.describe()
```

```python
fig= px.histogram(medical_df,
                  x='age',
                  marginal='box',
                  nbins=47,
                  color_discrete_sequence=['darkcyan'],
                  title='Distribution of Age')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig= px.histogram(medical_df,
                  x='bmi',
                  marginal='box',
                  nbins=47,
                  color_discrete_sequence=['lightslategrey'],
                  title='Distribution of BMI')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig= px.histogram(medical_df,
                  x='charges',
                  marginal='box',
                  color='smoker',
                  nbins=47,
                  color_discrete_sequence=['saddlebrown','peachpuff'],
                  title='Annual medical charges')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.scatter(medical_df,
                 x='age',
                 y='charges',
                 color='smoker',
                 opacity=0.8,
                 hover_data=['sex'],
                 title='Age vs. Charges')
fig.update_traces(marker_size=5)
fig.show()
```

```python
fig = px.scatter(medical_df,
                 x='bmi',
                 y='charges',
                 color='smoker',
                 opacity=0.8,
                 hover_data=['sex'],
                 title='Age vs. Charges')
fig.update_traces(marker_size=5)
fig.show()
```

```python
# Convert 'smoker' column to 1 for 'yes' and 0 for 'no'
medical_df['smoker'] = medical_df['smoker'].map({'yes': 1, 'no': 0})

# Now calculate the correlation
correlation = medical_df['charges'].corr(medical_df['smoker'])
print(correlation)
```

```python
medical_df['sex'] = medical_df['sex'].map({'female': 0, 'male': 1})
medical_df['smoker'] = medical_df['smoker'].map({'yes': 1, 'no': 0})
medical_df['region'] = pd.Categorical(medical_df['region']).codes  # Convert 'region' to numeric codes

# Select only numeric columns
numeric_df = medical_df.select_dtypes(include=['float64', 'int64'])

# Now calculate the correlation matrix
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```
