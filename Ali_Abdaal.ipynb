# **Python’s libraries like pandas, numpy, and faker can help create realistic datasets.**

```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# YouTube Metrics Data
youtube_metrics = {
    "date": pd.date_range(start="2023-01-01", end="2023-12-01", freq="MS"),
    "video_count": [random.randint(5, 10) for _ in range(12)],
    "subscriber_count": [random.randint(1000000 + i * 50000, 1000000 + (i + 1) * 50000) for i in range(12)],
    "total_views": [random.randint(200000000 + i * 5000000, 200000000 + (i + 1) * 5000000) for i in range(12)],
}

youtube_metrics_df = pd.DataFrame(youtube_metrics)

# Income Streams Data
income_streams = {
    "month": pd.date_range(start="2023-01-01", end="2023-12-01", freq="MS"),
    "adsense_income": [round(random.uniform(5000, 15000), 2) for _ in range(12)],
    "sponsorship_income": [round(random.uniform(10000, 30000), 2) for _ in range(12)],
    "affiliate_income": [round(random.uniform(5000, 20000), 2) for _ in range(12)],
    "courses_income": [round(random.uniform(20000, 50000), 2) for _ in range(12)],
}

income_streams_df = pd.DataFrame(income_streams)

# Video Performance Data
video_performance = {
    "video_id": range(1, 21),
    "title": [fake.sentence(nb_words=5) for _ in range(20)],
    "upload_date": pd.date_range(start="2023-01-01", periods=20, freq="7D"),
    "views": [random.randint(10000, 1000000) for _ in range(20)],
    "likes": [random.randint(1000, 50000) for _ in range(20)],
    "comments": [random.randint(50, 5000) for _ in range(20)],
}

video_performance_df = pd.DataFrame(video_performance)

# Expenses Data
expenses = {
    "month": pd.date_range(start="2023-01-01", end="2023-12-01", freq="MS"),
    "equipment_cost": [round(random.uniform(500, 2000), 2) for _ in range(12)],
    "team_salary": [round(random.uniform(10000, 50000), 2) for _ in range(12)],
    "ads_budget": [round(random.uniform(2000, 10000), 2) for _ in range(12)],
}

expenses_df = pd.DataFrame(expenses)

# Save datasets to CSV
youtube_metrics_df.to_csv("youtube_metrics.csv", index=False)
income_streams_df.to_csv("income_streams.csv", index=False)
video_performance_df.to_csv("video_performance.csv", index=False)
expenses_df.to_csv("expenses.csv", index=False)

print("Datasets created and saved as CSV files.")
```

# **Visualize Your Results**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file exported from your database
data = pd.read_csv('youtube_growth_.csv')

# Create a figure and axis to ensure proper configuration
plt.figure(figsize=(10, 6))  # Adjust figure size for better visibility

# Line plot for subscriber growth and total views
plt.plot(data['date'], data['subscriber_count'], label='Subscribers')
plt.plot(data['date'], data['total_views'], label='Total Views')

# Rotate x-axis labels to avoid overlap
plt.xticks(rotation=45, ha='right')

# Add legend, title, and labels
plt.legend()  # Ensure labels from `label` in `plt.plot` are used
plt.title("YouTube Growth Over Time")
plt.xlabel("Date")
plt.ylabel("Counts")

# Adjust layout to ensure elements fit nicely
plt.tight_layout()

# Show the plot
plt.show()

```
